```python
import numpy as np
from scipy.stats import norm

def bs_call(S, K, T, r, sigma):
    """
    BlackScholes price for a European call option.

    Parameters:
    -----------
    S     : float
            Current spot price of the underlying
    K     : float
            Strike price
    T     : float
            Time to expiry in years (e.g., 30 days  30/365)
    r     : float
            Annual riskfree rate (decimal, e.g. 0.07 for 7%)
    sigma : float
            Annual volatility (decimal, e.g. 0.20 for 20%)

    Returns:
    --------
    call_price : float
                 Theoretical BlackScholes call option price
    """
    # avoid division by zero
    if T <= 0 or sigma <= 0:
        return max(0.0, S - K * np.exp(-r * T))

    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    call_price = (S * norm.cdf(d1) 
                  - K * np.exp(-r * T) * norm.cdf(d2))
    return call_price
```

```python
def quick_backtest(df, price_col='Price', signal_col='X'):
    trades = []
    
    for i in range(len(df) - 1):  # stop at len-1 to avoid index error
        price_today = df.loc[i, price_col]
        signal_val = df.loc[i, signal_col]
        
        # Buy condition
        if price_today <   0.6*signal_val:
            entry_price = price_today
            exit_price = df.loc[i + 1, price_col]  # exit next day
            profit = exit_price - entry_price
            trades.append(profit)
    
    total_profit = sum(trades)
    num_trades = len(trades)
    avg_profit = total_profit / num_trades if num_trades else 0
    
    return {
        'total_profit': total_profit,
        'num_trades': num_trades,
        'avg_profit_per_trade': avg_profit
    }
```

```python
quick_backtest(data, price_col='SETTLE PRICE ', signal_col='BS_CALL')
```
